```python
import pandas as pd
# Displaying the TestInfo.csv as dataframe for knowing more insights
df1 = pd.read_csv('TestInfo.csv')
```

```python
df1.describe()
```

```python
# Displaying the TestResults.pickle as dataframe for knowing more insights
df2 = pd.read_pickle('TestResults.pickle')
```

```python
df2
```

```python
df2.describe()
```

```python
# Merging both dataframes based on common identifier 'TestId' for identifying underlying patterns in the data
df3 = pd.merge(df1, df2, on="TestId")
```

```python
df3
```

```python
df3.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
#Line graph to illustrate the correlation between the time and device type(ie'Device_0 or Device_1)
#From the data points its evident that the Device_0 is being used frequently, but we could also observe higer time is consuming by the Device_1 
time_0 = df3[df3['Device'] == 'Device_0']['Time (ms)']
time_1 = df3[df3['Device'] == 'Device_1']['Time (ms)']
plt.plot(time_0, label='Device_0')
plt.plot(time_1, label='Device_1')
plt.legend()
plt.show()
```

```python
#The bar graph also support the fact that  higer time is consuming by the Device_1 

# Grouping the data by Device and calculating the mean Time for each group
device_time_means = df3.groupby('Device')['Time (ms)'].mean()

# Creating a bar plot
plt.bar(device_time_means.index, device_time_means.values)
plt.xlabel('Device')
plt.ylabel('Mean Time (ms)')
plt.title('Mean Time by Device')

# Displaying the plot
plt.show()

```

```python
#The bar graph shows that higer peakmemory is ultised by the Device_0

# Grouping the data by Device and calculating the mean PeakMemory for each group
device_PeakMemory_means = df3.groupby('Device')['PeakMemory (MB)'].mean()

# Creating a bar plot
plt.bar(device_PeakMemory_means.index, device_PeakMemory_means.values)
plt.xlabel('Device')
plt.ylabel('Mean PeakMemory (MB)')
plt.title('Mean PeakMemory by Device')

# Displaying the plot
plt.show()

```

```python
#The scatter plots illustrate that tests with MobileNet tend to have shorter times compared to tests with AlexNet.

# Extract the values from the 'Time' and 'MLNetwork' columns
time = df3['Time (ms)']
ml_network = df3['MLNetwork']

# Create a scatter plot
plt.scatter(ml_network, time)

# Add labels to the x and y axes
plt.xlabel('ML Network')
plt.ylabel('Time (ms)')

# Show the plot
plt.show()

```

```python
#Illustrates the Distribution of CPU Frequency and it is evident that most of the device havce CPU Frequency : 1000 Mhz 

# Set up the figure and axis
fig, ax = plt.subplots()

# Plot the histogram
ax.hist(df3['CPUFrequency (MHz)'], bins=10)

# Add a title and x- and y-axis labels
ax.set_title('Distribution of CPU Frequency')
ax.set_xlabel('CPU Frequency (MHz)')
ax.set_ylabel('Frequency')

# Show the plot
plt.show()

```

```python
#Illustrate the Frequency disribution of Device Type

device = df3['Device']
frequency = df3['CPUFrequency (MHz)']

# Count the frequency of each device type
device_counts = device.value_counts()

# Plot the pie chart
device_counts.plot.pie(legend=True)

# Add a title
plt.title('Device Type and Frequency Distribution')

# Display the plot
plt.show()
```

```python
# Tests with optimisation enabled tend to have shorter times and lower memory usage 

# Grouping data by optimization flag
optimized = df3[df3['Optimised'] == 'Y']
not_optimized = df3[df3['Optimised'] == 'N']

# Calculating the mean time and memory for each group
optimized_time = optimized['Time (ms)'].mean()
not_optimized_time = not_optimized['Time (ms)'].mean()
optimized_memory = optimized['PeakMemory (MB)'].mean()
not_optimized_memory = not_optimized['PeakMemory (MB)'].mean()

# Set up bar plot
x = ['Optimized', 'Not Optimized']
y1 = [optimized_time, not_optimized_time]
y2 = [optimized_memory, not_optimized_memory]

fig, ax1 = plt.subplots()

# Plot time on left y-axis
color = 'tab:red'
ax1.set_xlabel('Optimization Flag')
ax1.set_ylabel('Time (ms)', color=color)
ax1.bar(x, y1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Plot memory on right y-axis
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Peak Memory (MB)', color=color)
ax2.bar(x, y2, color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Show plot
plt.show()
```
